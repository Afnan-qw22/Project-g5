```python
import os
import pandas as pd
import numpy as np
```

```python
books = pd.read_csv('books.csv', header=0)
books
```

**Omit some values of desired columns and row**

```python
books = books.drop('Unnamed: 0', axis = 1)
books = books.drop('Unnamed: 4', axis = 1)

books = books.drop(labels=[6,5,3], axis=0)
books
```

**Rename the columns names**

```python
columns = ['Unnamed: 1','Unnamed: 2','Unnamed: 3']

books.rename(columns = {columns[0]:'During of Period'}, inplace = True)
books.rename(columns = {columns[1]:'Number of Sukuk'}, inplace = True)
books.rename(columns = {columns[2]:"The Value(Million Riyals)"}, inplace = True)
books
```

**Consider the missing values**

```python
books['During of Period'] = books['During of Period'].fillna("No period")

books['Number of Sukuk'] = books['Number of Sukuk'].fillna(books['Number of Sukuk'].median())

books['The Value(Million Riyals)'] = books['The Value(Million Riyals)'].fillna(books['The Value(Million Riyals)'].median())

books
```

**Convert the data in the correct form**

```python
books['During of Period'] = books['During of Period'].apply(lambda x: str(x).split('\n')[1] if x != "No period" else x)

books['Number of Sukuk'] = books['Number of Sukuk'].astype('int')

books['The Value(Million Riyals)'] = books['The Value(Million Riyals)'].astype('float')

books
```
